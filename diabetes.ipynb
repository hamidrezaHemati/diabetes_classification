## 3.1 Handling Nan values
there are multiple options to do this task
 - deleting rows with Nan values using --> dropna()
 - replacing Nan values with proper values using --> fillna()
     - for categorical values the proper value to fill NaNs are most occurance value in column
     - for numerical vales the proper value to fill NaNs are mean() of the column

```python
pdf = df #pdf stands for preprocessed data frame
pdf.isnull().sum()
```

```python
pdf['Income'].value_counts()
```

```python
def one_hot_encoding(column_name, pdf):
    one_hot_encoded_df = pd.get_dummies(pdf[column_name])
    new_names = []
    for col in list(one_hot_encoded_df.columns):
        new_names.append(column_name + "_" + col)
    one_hot_encoded_df.columns = new_names
    pdf = pd.concat([pdf,one_hot_encoded_df],axis=1)
    
```

```python
for column_name in ["general_health", "sex", "education", "income"]:
    one_hot_encoded_df = pd.get_dummies(pdf[column_name])
    new_names = []
    for col in list(one_hot_encoded_df.columns):
        new_names.append(column_name + "_" + col)
    one_hot_encoded_df.columns = new_names
    pdf = pd.concat([pdf,one_hot_encoded_df],axis=1)
pdf
```

# 4. Hyperparameter setting

```python
xgb.XGBClassifier().get_params()
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import roc_auc_score
from xgboost import XGBClassifier
```

```python
def my_roc_auc_score(classifier, X, y): return roc_auc_score(y, classifier.predict_proba(X)[:,1])

parameters = {
    'learning_rate' : [0.02,0.05,0.1,0.3],
    'max_depth' : [2,3,4],
    'n_estimators' : [100,200,300],
    'colsample_bytree' : [0.8,1]
}

estimator = XGBClassifier(
    objective= 'binary:logistic',
    nthread=4,
    seed=42
)
grid_search = GridSearchCV(
    estimator=estimator,
    param_grid=parameters,
    scoring = my_roc_auc_score,
    cv = 3
    )

grid_search.fit(X_train, y_train) 
print("Best score: %f using %s" % (grid_search.best_score_, grid_search.best_params_))
```

```python
grid_predict = grid_search.predict(X_test) 
   
# print classification report 
print(classification_report(y_test, grid_predict))
```
